### Часть 1: MNIST
Обучите полносвязную нейронную сеть с архитектурой 784-100-100-10 и сигмоидой в качестве нелинейности. Какую точность классификации удалось получить? Нарисуйте график сходимости на обучающей и тестовой выборках. В качестве темпа обучения (learning rate) возьмите 0.01

```python
def train(network, epochs, learning_rate, plot=True,
          verbose=True, loss=None):
    loss = loss or NLLLoss()
    train_loss_epochs = []
    test_loss_epochs = []
    train_accuracy_epochs = []
    test_accuracy_epochs = []
    try:
        for epoch in range(epochs):
            losses = []
            accuracies = []
            for X, y in train_loader:
                X = X.view(X.shape[0], -1).numpy()
                y = y.numpy()
                prediction = network.forward(X)
                loss_batch = loss.forward(prediction, y)
                losses.append(loss_batch)
                dLdx = loss.backward()
                network.backward(dLdx)
                network.step(learning_rate)
                accuracies.append((np.argmax(prediction, 1)==y).mean())
            train_loss_epochs.append(np.mean(losses))
            train_accuracy_epochs.append(np.mean(accuracies))
            losses = []
            accuracies = []    
            for X, y in test_loader:
                X = X.view(X.shape[0], -1).numpy()
                y = y.numpy()
                prediction = network.forward(X)
                loss_batch = loss.forward(prediction, y)
                losses.append(loss_batch)
                accuracies.append((np.argmax(prediction, 1)==y).mean())
            test_loss_epochs.append(np.mean(losses))
            test_accuracy_epochs.append(np.mean(accuracies))
            clear_output(True)
            if verbose:
                sys.stdout.write('\rEpoch {0}... (Train/Test) NLL: {1:.3f}/{2:.3f}\tAccuracy: {3:.3f}/{4:.3f}'.format(
                            epoch, train_loss_epochs[-1], test_loss_epochs[-1],
                            train_accuracy_epochs[-1], test_accuracy_epochs[-1]))
            if plot:
                plt.figure(figsize=(12, 5))
                plt.subplot(1, 2, 1)
                plt.plot(train_loss_epochs, label='Train')
                plt.plot(test_loss_epochs, label='Test')
                plt.xlabel('Epochs', fontsize=16)
                plt.ylabel('Loss', fontsize=16)
                plt.legend(loc=0, fontsize=16)
                plt.grid('on')
                plt.subplot(1, 2, 2)
                plt.plot(train_accuracy_epochs, label='Train accuracy')
                plt.plot(test_accuracy_epochs, label='Test accuracy')
                plt.xlabel('Epochs', fontsize=16)
                plt.ylabel('Accuracy', fontsize=16)
                plt.legend(loc=0, fontsize=16)
                plt.grid('on')
                plt.show()
    except KeyboardInterrupt:
        pass
    return train_loss_epochs, \
           test_loss_epochs, \
           train_accuracy_epochs, \
           test_accuracy_epochs
```

```python
network = NeuralNetwork([
    Linear(784, 50), Sigmoid(), # 28 * 28
    Linear(50, 25), Sigmoid(),
    Linear(25, 10)
])
loss = NLLLoss()
```

```python
tr_s, ts_s, tr_ac_s, ts_ac_s = train(network, 20, 0.01)
```

### Часть 2: Нелинейности

Обучите нейронную сеть с другими нелинейностями: ReLU и ELU. Сравните скорости сходимости и качество классификации с различными функциями активации.

```python
class ReLU:
    def __init__(self):
        pass

    def forward(self, X):
        '''
        Passes objects through this layer.
        X is np.array of size (N, d)
        '''
        self.X = X
        return np.maximum(X, 0)

    def backward(self, dLdy):
        '''
        1. Compute dLdx.
        2. Return dLdx
        '''
        return (self.X > 0) * dLdy

    def step(self, learning_rate):
        pass
```

```python
class ELU:
    '''
    ELU(x) = x, x > 0; a*(e^x - 1), x <= 0
    '''
    
    def __init__(self, a=1):
        self.a = a

    def forward(self, X):
        '''
        Passes objects through this layer.
        X is np.array of size (N, d)
        '''
        self.X = X
        return X * (X > 0) + self.a * (np.exp(X) - 1) * (X <= 0)
    
    def backward(self, dLdy):
        '''
        1. Compute dLdx.
        2. Return dLdx
        '''
        X = self.X
        dydX = (X > 0) + self.a * np.exp(X) * (X <= 0)
        return dLdy*dydX

    def step(self, learning_rate):
        pass
```

### HW1 (1 балл): Реализовать Tanh

Обучите нейронную сеть с другими нелинейностями: ReLU и ELU. Сравните скорости сходимости и качество классификации с различными функциями активации.

```python
class Tanh:
    def __init__(self):
        pass

    def forward(self, X):
        '''
        Passes objects through this layer.
        X is np.array of size (N, d)
        '''
        self.X = X
        self.exps = np.exp(-2 * X)
        return 2 / (1 + self.exps) - 1

    def backward(self, dLdy):
        '''
        1. Compute dLdx.
        2. Return dLdx
        '''
        return dLdy * 4 * self.exps / (1 + self.exps) ** 2

    def step(self, learning_rate):
        pass
```

### Часть 4: Аугментация (HW, 2 балла)

Какой прирост дают эти аугментации вместе и по отдельности

```python
from skimage.transform import rotate
```

```python
def train_aug(network, epochs, learning_rate,
              rotates = 0, moves = 0, noise = 0,
              plot=True, verbose=True, loss=None):
    """
    One hidden layer M x N.
    Rotate: in (-rotate, rotate) degrees.
    Move = (m, ax): in (-m, m) by axis ax.
    Noise: in (-noise, noise).
    """
    loss = loss or NLLLoss()
    train_loss_epochs = []
    test_loss_epochs = []
    train_accuracy_epochs = []
    test_accuracy_epochs = []

    try:
        for epoch in range(epochs):
            losses = []
            accuracies = []
            for X, y in train_loader:
                X = X.view(X.shape[0], -1).numpy()
                y = y.numpy()
                
                for sample in X:
                    if rotates != 0:
                        angle = rotates * (-1 + 2 * np.random.random())
                        sample = rotate(sample.reshape(28, 28), angle)
                    if moves != 0:
                        mv = np.random.randint(low=-moves[0], high=moves[0])
                        sample = np.roll(sample, mv, axis=moves[1])
                    if noise != 0:
                        ns = noise * (-1 + 2 * np.random.random(size=sample.shape))
                        sample += ns
                    X = np.vstack((X, sample.reshape(X.shape[1:])))
                y = np.hstack((y, y))

                prediction = network.forward(X)
                loss_batch = loss.forward(prediction, y)
                losses.append(loss_batch)
                dLdx = loss.backward()
                network.backward(dLdx)
                network.step(learning_rate)
                accuracies.append((np.argmax(prediction, 1)==y).mean())
            train_loss_epochs.append(np.mean(losses))
            train_accuracy_epochs.append(np.mean(accuracies))
            losses = []
            accuracies = []    
            for X, y in test_loader:
                X = X.view(X.shape[0], -1).numpy()
                y = y.numpy()
                prediction = network.forward(X)
                loss_batch = loss.forward(prediction, y)
                losses.append(loss_batch)
                accuracies.append((np.argmax(prediction, 1)==y).mean())
            test_loss_epochs.append(np.mean(losses))
            test_accuracy_epochs.append(np.mean(accuracies))
            clear_output(True)
            if verbose:
                sys.stdout.write("""\rRotate = {5}, Move = {6}, Noise = {7}
                                    \nEpoch {0}... (Train/Test) NLL: {1:.3f}/{2:.3f} \tAccuracy: {3:.3f}/{4:.3f}""".format(
                                 epoch, train_loss_epochs[-1], test_loss_epochs[-1],
                                 train_accuracy_epochs[-1], test_accuracy_epochs[-1],
                                 rotates, moves, noise))
            if plot:
                plt.figure(figsize=(12, 5))
                plt.subplot(1, 2, 1)
                plt.plot(train_loss_epochs, label='Train')
                plt.plot(test_loss_epochs, label='Test')
                plt.xlabel('Epochs', fontsize=16)
                plt.ylabel('Loss', fontsize=16)
                plt.legend(loc=0, fontsize=16)
                plt.grid('on')
                plt.subplot(1, 2, 2)
                plt.plot(train_accuracy_epochs, label='Train accuracy')
                plt.plot(test_accuracy_epochs, label='Test accuracy')
                plt.xlabel('Epochs', fontsize=16)
                plt.ylabel('Accuracy', fontsize=16)
                plt.legend(loc=0, fontsize=16)
                plt.grid('on')
                plt.show()
    except KeyboardInterrupt:
        pass
    return train_loss_epochs, \
           test_loss_epochs, \
           train_accuracy_epochs, \
           test_accuracy_epochs
```

```python
transform = transforms.Compose([
                       transforms.ToTensor(),
                       transforms.Normalize((0.1307,), (0.3081,))
                    ])
train_dataset = MNIST('.', train=True, download=True, transform=transform)
test_dataset = MNIST('.', train=False, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
def aug_sample(X, rotates = 0, moves = 0, noise = 0):
    X = X.view(X.shape[0], -1).numpy()
    for sample in X:
        if rotates != 0:
            angle = rotates * (-1 + 2 * np.random.random())
            sample = rotate(sample.reshape(28, 28), angle)
        if moves != 0:
            mv = np.random.randint(low=-moves[0], high=moves[0])
            sample = np.roll(sample, mv, axis=moves[1])
        if noise != 0:
            ns = noise * (-1 + 2 * np.random.random(size=sample.shape))
            sample += ns
    return sample

plt.figure(figsize=(9, 6))
X, y = next(iter(train_loader))
for i in range(5):
    plt.subplot(4, 5, i+1)
    plt.imshow(aug_sample(X[i], rotates=15, moves=0, noise=0).reshape(28, 28), cmap=plt.cm.Greys_r)
    plt.title(int(y[i]))
    plt.axis('off')
for i in range(5):
    plt.subplot(4, 5, i+6)
    plt.imshow(aug_sample(X[i], rotates=0, moves=(3, 0), noise=0).reshape(28, 28), cmap=plt.cm.Greys_r)
    plt.title(int(y[i]))
    plt.axis('off')
for i in range(5):
    plt.subplot(4, 5, i+11)
    plt.imshow(aug_sample(X[i], rotates=0, moves=0, noise=0.5).reshape(28, 28), cmap=plt.cm.Greys_r)
    plt.title(int(y[i]))
    plt.axis('off')
for i in range(5):
    plt.subplot(4, 5, i+16)
    plt.imshow(aug_sample(X[i], rotates=15, moves=(3, 0), noise=0.5).reshape(28, 28), cmap=plt.cm.Greys_r)
    plt.title(int(y[i]))
    plt.axis('off')
```

```python
def test_augmented():
    result = {}
    for r in [0, 15]:
        for m in [0, (3, 0)]:
            for n in [0, 0.6]:
                network = NeuralNetwork([
                    Linear(784, 50), Tanh(),
                    Linear(50, 25), Tanh(),
                    Linear(25, 10)
                ])
                loss = NLLLoss()
                tr, ts, tr_ac, ts_ac = train_aug(network, 10, 0.01, 
                                                 rotates=r, moves=m, noise=n,
                                                 plot=False, verbose=True)
                result["(r={}, m={}, n={})".format(r, m, n)] = ts_ac[-1]
    return result
```

```python
result = test_augmented()
```
